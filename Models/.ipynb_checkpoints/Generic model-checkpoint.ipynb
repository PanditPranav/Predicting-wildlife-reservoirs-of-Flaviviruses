## Holdout dataset and cross validation

```r

library(caTools)
model.data$spl=sample.split(model.data$TotalBinary,SplitRatio=0.8)
train=subset(model.data, model.data$spl==TRUE)
test=subset(model.data, model.data$spl==FALSE)
nrow(train)
nrow(test)
table(model.data$TotalBinary)
table(train$TotalBinary)
table(test$TotalBinary)
```

```r
set.seed(100)
Group1.total.tc5.lr0005 <- gbm.step(data = train,
                                   gbm.x = predictorsNames,
                                   gbm.y = outcomeName,
                                   family = "bernoulli",
                                   tree.complexity = 5,
                                   learning.rate = 0.001,
                                   bag.fraction = 0.5, 
                                   cv = 10, 
                                   max.trees = 20000)
```

```r
print(Group1.total.tc5.lr0005)
```

```r
preds2 <- predict.gbm(Group1.total.tc5.lr0005, 
                     test[,predictorsNames],
                     n.trees=Group1.total.tc5.lr0005$gbm.call$best.trees,
                     type="response")

calc.deviance(obs=test$TotalBinary, pred=preds2, calc.mean=TRUE)
```

```r
d <- cbind(test$TotalBinary, preds2)
pres <- d[d[,1]==1, 2]
abs <- d[d[,1]==0, 2]
e <- evaluate(p=pres, a=abs)
e
```

```r
preds2 <- predict.gbm(Group1.total.tc5.lr0005, 
                     model.data[,predictorsNames],
                     n.trees=Group1.total.tc5.lr0005$gbm.call$best.trees,
                     type="response")
```

```r
model.data$PredictionProbability  = preds2
write.csv(model.data, file = "Outputs2/Predictions_R_Generic2.csv")
```

```r
ranks = relative.influence(Group1.total.tc5.lr0005, scale. = 'TRUE', sort. = 'TRUE')
a = as.data.frame(ranks, col.names = , optional = FALSE)
a = setDT(a, keep.rownames = TRUE)[]
colnames(a) = c('Variable','influence')
b = a
for (i in b$Variable){
    margianleffects = plot.gbm(Group1.total.tc5.lr0005, i.var = i,n.plots=3, 
                           write.title = T, return.grid  = T,
                           show.contrib = T,y.label="Marginal effect on Host")
    filename = paste("Outputs2/PartialDependence2/", i,'_GenericFlavi','.csv', sep="")
    write.csv(margianleffects, file = filename)    
}
```

```r
Interaction = plot.gbm(Group1.total.tc5.lr0005, c('centroidX_log','centroidY'),  
                           write.title = T, return.grid  = T,
                           show.contrib = T)
    filename = 'Outputs2/PartialDependence2/iteraction_centroid_X_Y_GenericFlavi.csv'
    write.csv(Interaction, file = filename) 

#plot.gbm(Generic.total.tc5.lr0005, c('centroidX_log','centroidY'), col= heat.colors(8, alpha=0.05))
```
